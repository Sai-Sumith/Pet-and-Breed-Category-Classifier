```python
#importing required libraries
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

```python
#Loading training and test dataset
train_df=pd.read_csv('/home/sumith/Desktop/dog_train.csv')
test_df=pd.read_csv('/home/sumith/Desktop/dog_test.csv')
```

```python
#Checking first rows of train dataset
train_df.head()
```

```python
#Checking first rows of test dataset
test_df.head()
```

```python
#Checking datatypes of dataset
train_df.dtypes
```

```python
#Checking Correlation between different Variables
train_df.corr()
```

```python
#Mean Encoding 'color_type' categorical variable for train data.
mean=train_df.groupby(['color_type'])['X1'].mean() 
Mean_encoded_subject =train_df.groupby(['color_type'])['X1'].mean().to_dict() 
train_df['color_type'] =train_df['color_type'].map(Mean_encoded_subject) 
```

```python
#Mean Encoding 'color_type' categorical variable for test data.
mean=test_df.groupby(['color_type'])['X1'].mean() 
Mean_encoded_subject =test_df.groupby(['color_type'])['X1'].mean().to_dict() 
test_df['color_type'] =test_df['color_type'].map(Mean_encoded_subject)
```

```python
#Checking for null values in train dataset
train_df.isnull()
```

```python
#Checking for null values in test dataset
test_df.isnull()
```

```python
#Filling empty values in test and training dataset
train_df['condition'].fillna(value=1,inplace=True)
test_df['condition'].fillna(value=1,inplace=True)
```

```python
#Min-Max normalising the training dataset(Using only highly correlated variables).
train_df['X1']=(train_df['X1']-train_df['X1'].min())/(train_df['X1'].max()-train_df['X1'].min())
train_df['X2']=(train_df['X2']-train_df['X2'].min())/(train_df['X2'].max()-train_df['X2'].min())
train_df['condition']=(train_df['condition']-train_df['condition'].min())/(train_df['condition'].max()-train_df['condition'].min())
train_df['color_type']=(train_df['color_type']-train_df['color_type'].min())/(train_df['color_type'].max()-train_df['color_type'].min())
```

```python
#Min-Max normalising the test dataset(Using only highly correlated variables).
test_df['X1']=(test_df['X1']-test_df['X1'].min())/(test_df['X1'].max()-test_df['X1'].min())
test_df['X2']=(test_df['X2']-test_df['X2'].min())/(test_df['X2'].max()-test_df['X2'].min())
test_df['condition']=(test_df['condition']-test_df['condition'].min())/(test_df['condition'].max()-test_df['condition'].min())
test_df['color_type']=(test_df['color_type']-test_df['color_type'].min())/(test_df['color_type'].max()-test_df['color_type'].min())
```

```python
#Splitting into dependent and independent variables
X_train = train_df[['condition', 'X1','X2','color_type']]
y_train1 = train_df['breed_category']
y_train2=train_df['pet_category']
X_test = test_df[[ 'condition', 'X1','X2','color_type']] 
```

```python
#Initialising Decision Tree Classifier
dtree1 = DecisionTreeClassifier()
dtree2 = DecisionTreeClassifier()
```

```python
#Fit into training dataset
dtree1.fit(X_train,y_train1)
dtree2.fit(X_train,y_train2)
```

```python
#Predicting the test dataset
pred1=dtree1.predict(X_test)
pred2=dtree2.predict(X_test)
```

```python
#Creating submission dataframe for submitting csv file
sub=pd.DataFrame()
sub['pet_id']=test_df['pet_id']
```

```python
#Filling target variables and generating csv file
sub['breed_category']=pred1
sub['pet_category']=pred2
sub.to_csv('/home/sumith/Desktop/dog_pred.csv',index=False)
```

```python
#Using Random Forest Classifier for first target variable
clf1=RandomForestClassifier()
clf1.fit(X_train,y_train1)
pred1= clf1.predict(X_test)
```

```python
#Using Random Forest Classifier for second target variable
clf2=RandomForestClassifier()
clf2.fit(X_train,y_train2)
pred2= clf2.predict(X_test)
```
